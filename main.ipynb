```python
# Encapsulamento

class Conta:
    def __init__(self, saldo):
        self._saldo = saldo

    def depostirar(self, valor):
        self._saldo += valor

    def sacar(self, valor):
        self._saldo -= valor

    def mostrar_saldo(self):
        print(f"Saldo Atual: R$ {self._saldo:.2f}")

conta = Conta(5000)
conta.mostrar_saldo()
```

```python
# Propiedades

class Foo:
    def __init__(self, x=None):
        self._x = x

    @property
    def x(self):
        return self._x or 0

    @x.setter
    def x(self, valor):
        self._x += valor

    @x.deleter
    def x(self):
        self._x -= 1

name = Foo(10)

print(name.x)
name.x = 10
print(name.x)
```

```python
# Polimorfismo

class Passaro:
    def voar(self):
        print("Voando...")

class Pardal(Passaro):
    def voar(self):
        return super().voar()

class Avestrus(Passaro):
    def voar(self):
        print("Avestrus não voa")

def plano_de_voo(passaro):
    passaro.voar()

p1 = Pardal()
p2 = Avestrus()

plano_de_voo(p1)
plano_de_voo(p2)
```

```python
# Atributo clase instancia

class Estudante:
    escola = "DIO"

    def __init__(self, nome, matricula):
        self.nome = nome
        self.matricula = matricula

    def __str__(self):
        return f"{self.__class__.__name__}: {', '.join([f'{chave}={valor}' for chave, valor in self.__dict__.items()])}"

def mostrar_valores(*objs):
    for obj in objs:
        print(obj)
        
aluno1 = Estudante("João Pedro", 14837)
aluno2 = Estudante("Pedro João", 73841)

mostrar_valores(aluno1)
mostrar_valores(aluno2)

aluno1.escola = 'Python'
print(aluno1)
```

```python
# Classes Abstratas

from abc import ABC, abstractmethod

class Veiculo(ABC):
    @abstractmethod
    def acelerar(self):
        pass
    
    @abstractmethod
    def frear(self):
        pass

class Carro(Veiculo):
    def acelerar(self):
        print("Carro acelerando...")
    
    def frear(self):
        print("Carro freando...")

class Bicicleta(Veiculo):
    def acelerar(self):
        print("Bicicleta acelerando...")
    
    def frear(self):
        print("Bicicleta freando...")

# car = Veiculo()  # Isso levantaria um erro porque Veiculo é uma classe abstrata
carro = Carro()
carro.acelerar()  # Saída: Carro acelerando...
carro.frear()     # Saída: Carro freando...
```
